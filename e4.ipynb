```python
import good_sorts
from bad_sorts import create_random_list
import time
import matplotlib.pyplot as plt

# Author: Emre Bozkurt
```

```python
# Times for sorts: merge, quick, and heap sorts
lengths = [n * 1000 for n in range (1, 11)]  # from 1000 to 10000 in steps of 1000
merge_times = []
quick_times = []
heap_times = []

for n in lengths:
    lst = create_random_list(n, 512)

    start = time.time()
    good_sorts.mergesort(lst.copy())
    end = time.time()
    merge_times.append(end - start)

    start = time.time()
    good_sorts.quicksort(lst.copy())
    end = time.time()
    quick_times.append(end - start)

    start = time.time()
    good_sorts.heapsort(lst.copy())
    end = time.time()
    heap_times.append(end - start)

# Plotting the results
plt.plot(lengths, merge_times, label='Merge Sort')
plt.plot(lengths, quick_times, label='Quick Sort')
plt.plot(lengths, heap_times, label='Heap Sort')
plt.xlabel('List Length')
plt.ylabel('Time (seconds)')
plt.title('Sorting Algorithm Times')
plt.legend()
plt.show()
```
