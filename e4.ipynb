```python
import good_sorts
from bad_sorts import create_random_list
import time
import matplotlib.pyplot as plt

# Author: Emre Bozkurt
```

```python
# Times for sorts: merge, quick, and heap sorts
lengths = [n * 1000 for n in range (1, 11)]  # from 1000 to 10000 in steps of 1000
runs = 10
merge_times = [0] * len(lengths)
quick_times = [0] * len(lengths)
heap_times = [0] * len(lengths)

for i in range(runs):
    for idx, n in enumerate(lengths):
        lst = create_random_list(n, 512)

        start = time.time()
        good_sorts.mergesort(lst.copy())
        end = time.time()
        merge_times[idx] += end - start

        start = time.time()
        good_sorts.quicksort(lst.copy())
        end = time.time()
        quick_times[idx] += end - start

        start = time.time()
        good_sorts.heapsort(lst.copy())
        end = time.time()
        heap_times[idx] += end - start

m_times = [t / runs for t in merge_times]
q_times = [t / runs for t in quick_times]
h_times = [t / runs for t in heap_times]

# Plotting the results
plt.plot(lengths, m_times, label='Merge Sort')
plt.plot(lengths, q_times, label='Quick Sort')
plt.plot(lengths, h_times, label='Heap Sort')
plt.xlabel('List Length')
plt.ylabel('Time (seconds)')
plt.title('Sorting Algorithm Times')
plt.legend()
plt.show()
```
