```python
import bad_sorts
import matplotlib.pyplot as plt
import time
# Author: Emre Bozkurt
```

```python
# Times for bubble, insertion, and selection sorts
lengths = [n * 1000 for n in range (1, 11)]  # from 1000 to 10000 in steps of 1000
runs = 3

bubble_times = [0] * len(lengths)
insertion_times = [0] * len(lengths)
selection_times = [0] * len(lengths)

# Average times over multiple runs
for i in range(runs):
    for idx, n in enumerate(lengths):
        lst = bad_sorts.create_random_list(n, 512)

        start = time.time()
        bad_sorts.bubble_sort(lst.copy())
        end = time.time()
        bubble_times[idx] += end - start

        start = time.time()
        bad_sorts.insertion_sort(lst.copy())
        end = time.time()
        insertion_times[idx] += end - start

        start = time.time()
        bad_sorts.selection_sort(lst.copy())
        end = time.time()
        selection_times[idx] += end - start

b_times = [t / runs for t in bubble_times]
i_times = [t / runs for t in insertion_times]
s_times = [t / runs for t in selection_times]


# plot all 3 algorithms into one graph
plt.plot(lengths, b_times, label='Bubble Sort')
plt.plot(lengths, i_times, label='Insertion Sort')
plt.plot(lengths, s_times, label='Selection Sort')
plt.xlabel('List Length')
plt.ylabel('Time (seconds)')
plt.title('Sorting Algorithm Times')
plt.legend()
plt.show()
```
