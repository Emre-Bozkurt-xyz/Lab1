```python
import bad_sorts
import matplotlib.pyplot as plt
import time
```

```python
# Times for bubble, insertion, and selection sorts
lengths = [n for n in range(100, 2001, 100)]  # from 100 to 2000 in steps of 100

bubble_times = []
insertion_times = []
selection_times = []

for n in lengths:
    lst = bad_sorts.create_random_list(n, 512)

    start = time.time()
    bad_sorts.bubble_sort(lst.copy())
    end = time.time()
    bubble_times.append(end - start)

    start = time.time()
    bad_sorts.insertion_sort(lst.copy())
    end = time.time()
    insertion_times.append(end - start)

    start = time.time()
    bad_sorts.selection_sort(lst.copy())
    end = time.time()
    selection_times.append(end - start)

# plot all 3 algorithms into one graph
plt.plot(lengths, bubble_times, label='Bubble Sort')
plt.plot(lengths, insertion_times, label='Insertion Sort')
plt.plot(lengths, selection_times, label='Selection Sort')
plt.xlabel('List Length')
plt.ylabel('Time (seconds)')
plt.title('Sorting Algorithm Times')
plt.legend()
plt.show()
```
