```python

import bad_sorts
import matplotlib.pyplot as plt
import time

# Author: Sreyo Biswas
```

```python
#Parameters for following experiements below
length = 5000
swaps = [0, 10, 50, 100, 500, 1000, 2000, 5000, 10000, 20000]
max_value = 512
runs = 5
bubble_times = [0] * len(swaps)
insertion_times = [0] * len(swaps)
selection_times = [0] * len(swaps)
```

```python
#Times for bubble, insertion and selection sorts on near sorted lists
for i in range(runs):
    for index, numSwaps in enumerate(swaps):
        #Create a near sorted list with given number of swaps
        L = bad_sorts.create_near_sorted_list(length, max_value, numSwaps)

        #Record times for bubble sort
        start = time.time()
        bad_sorts.bubble_sort(L.copy())
        end = time.time()
        #Accumulate total time
        bubble_times[index] += end - start

        #Record times for insertion sort
        start = time.time()
        bad_sorts.insertion_sort(L.copy())
        end = time.time()
        #Accumulate total time
        insertion_times[index] += end - start

        #Record times for selection sort
        start = time.time()
        bad_sorts.selection_sort(L.copy())
        end = time.time()
        #Accumulate total time
        selection_times[index] += end - start
#Get averages by dividing by number of runs
bubble_times = [t / runs for t in bubble_times]
insertion_times = [t / runs for t in insertion_times]
selection_times = [t / runs for t in selection_times]
```

```python
#Graph of final results displaying swaps vs time for each sorting algorithm
plt.plot(swaps, bubble_times, label = "Bubble Sort")
plt.plot(swaps, insertion_times, label= "Insertion Sort")
plt.plot(swaps, selection_times, label = "Selection Sort")
plt.xlabel("Number of Random Swaps")
plt.ylabel("Time(s)")
plt.title("Swaps vs Time(s) for Bad Sorting Algorithms")
plt.legend()
plt.show()
```
