```python

import bad_sorts
import matplotlib.pyplot as plt
import time

# Author: Sreyo Biswas
```

```python
#Parameters for following experiements below
length = 5000
swaps = [0, 10, 50, 100, 500, 1000, 2000, 5000, 10000, 20000]
max_value = 512
bubble_times = []
insertion_times = []
selection_times = []
```

```python
#Getting time for bad Bubble Sort implementation 
for numSwaps in swaps:
    #Create a near sorted list with given number of swaps
    L = bad_sorts.create_near_sorted_list(length, max_value, numSwaps)
    #Record times for sorting
    start = time.time()
    bad_sorts.bubble_sort(L.copy())
    end = time.time()
    
    #Append final calculated total time
    bubble_times.append(end - start)
```

```python
#Getting time for bad Insertion Sort implementation
for numSwaps in swaps:
    #Create a near sorted list with given number of swaps
    L = bad_sorts.create_near_sorted_list(length, max_value, numSwaps)
    #Record times for sorting
    start = time.time()
    bad_sorts.insertion_sort(L.copy())
    end = time.time()
    
    #Append final calculated total time
    insertion_times.append(end - start)
```

```python
#Getting time for bad Selection Sort implementation
for numSwaps in swaps:
    #Create a near sorted list with given number of swaps
    L = bad_sorts.create_near_sorted_list(length, max_value, numSwaps)
    #Record times for sorting
    start = time.time()
    bad_sorts.selection_sort(L.copy())
    end = time.time()
    
    #Append final calculated total time
    selection_times.append(end - start)
```

```python
#Graph of final results displaying swaps vs time for each sorting algorithm
plt.plot(swaps, bubble_times, label = "Bubble Sort")
plt.plot(swaps, insertion_times, label= "Insertion Sort")
plt.plot(swaps, selection_times, label = "Selection Sort")
plt.xlabel("Number of Random Swaps")
plt.ylabel("Time(s)")
plt.title("Swaps vs Time(s) for Bad Sorting Algorithms")
plt.legend()
plt.show()
```
