```python
import good_sorts
import bad_sorts
import matplotlib.pyplot as plt
import time

# Author: Sreyo Biswas

```

```python
#Parameters for following experiments below
length = 5000
max_value = 512
swaps = [10, 50, 100, 500, 1000, 2000, 5000]
quickSort_times = []
heapSort_times = []
mergeSort_times = []
```

```python
#Time for Quick Sort algorithm
for numSwaps in swaps:
    #Create near sorted list
    L = bad_sorts.create_near_sorted_list(length, max_value, numSwaps)
    #Record times for the quick sort algo
    start = time.time()
    good_sorts.quicksort(L)
    end = time.time()

    #Append final calculated total time
    quickSort_times.append(end - start)
```

```python
#Time for Merge Sort algorithm
for numSwaps in swaps:
    #Create near sorted list
    L = bad_sorts.create_near_sorted_list(length, max_value, numSwaps)
    #Record times for the merge sort algo
    start = time.time()
    good_sorts.mergesort(L.copy())
    end = time.time()

    #Append final calculated total time
    mergeSort_times.append(end - start)
```

```python
#Time for Heap Sort algorithm
for numSwaps in swaps:
    #Create near sorted list
    L = bad_sorts.create_near_sorted_list(length, max_value, numSwaps)
    #Record times for the heap sort algo
    start = time.time()
    good_sorts.heapsort(L)
    end = time.time()

    #Append final calculated total time
    heapSort_times.append(end - start)
```

```python
#Graph of final results displaying swaps vs time(s) for each "good" sorting algorithm
plt.plot(swaps, quickSort_times, label= "Quick Sort")
plt.plot(swaps, mergeSort_times, label = "Merge Sort")
plt.plot(swaps, heapSort_times, label = "Heap Sort")
plt.xlabel("Number of Random Swaps")
plt.ylabel("Time(s)")
plt.title("Swaps vs Time(s) for `Good` Sorting Algorithms")
plt.legend()
plt.show()


```
