```python
import good_sorts
import bad_sorts
import matplotlib.pyplot as plt
import time

# Author: Sreyo Biswas

```

```python
#Parameters for following experiments below
length = 5000
max_value = 100000
swaps = [10, 25, 50, 100, 200, 300, 500, 1000, 2000, 5000]
runs = 10
quickSort_times = [0] * len(swaps)
mergeSort_times = [0] * len(swaps)
heapSort_times = [0] * len(swaps)
```

```python
#Times for Quick, Merge and Heap sort on near sorted lists
for i in range(runs):
    for index, numSwaps in enumerate(swaps):
        #Create near sorted list
        L = bad_sorts.create_near_sorted_list(length, max_value, numSwaps)
        #Record times for the quick sort algo
        start = time.time()
        good_sorts.quicksort(L.copy())
        end = time.time()
        #Accumulate total time
        quickSort_times[index] += (end - start)

        #Record times for the merge sort algo
        start = time.time()
        good_sorts.mergesort(L.copy())
        end = time.time()
        #Accumulate total time
        mergeSort_times[index] += (end - start)
        
        #Record times for the heap sort algo
        start = time.time()
        good_sorts.heapsort(L.copy())
        end = time.time()
        #Accumulate total time
        heapSort_times[index] += (end - start)
quickSort_times = [t / runs for t in quickSort_times]
mergeSort_times = [t / runs for t in mergeSort_times]
heapSort_times = [t / runs for t in heapSort_times]
```

```python
#Graph of final results displaying swaps vs time(s) for each "good" sorting algorithm
plt.plot(swaps, quickSort_times, label= "Quick Sort")
plt.plot(swaps, mergeSort_times, label = "Merge Sort")
plt.plot(swaps, heapSort_times, label = "Heap Sort")
plt.xlabel("Number of Random Swaps")
plt.ylabel("Time(s)")
plt.title("Swaps vs Time(s) for `Good` Sorting Algorithms")
plt.legend()
plt.show()


```
