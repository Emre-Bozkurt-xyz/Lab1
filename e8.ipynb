```python
import bad_sorts
import time
import good_sorts
import matplotlib.pyplot as plt

# Author: Sreyo Biswas
```

```python
#Parameters for following experiments below
length = [5, 10, 15, 20, 25, 30, 40, 50, 60, 75, 100, 150, 200, 300, 500]
runs = 1000 
max_value = 10000
quickSort_times = [0] * len(length)
mergeSort_times = [0] * len(length)
insertionSort_times = [0] * len(length)
```

```python
for i in range(runs):
    for index, n in enumerate(length):
            #Create a random list of given length and max value
            L = bad_sorts.create_random_list(n, max_value)

            #Record times for insertion sort
            start = time.time()
            #Note : The optimized version for insertion sort will be used for comparison
            bad_sorts.insertion_sort2(L.copy())
            end = time.time()
            #Accumulate total time
            insertionSort_times[index] += (end - start)

            #Record times for merge sort
            start = time.time()
            good_sorts.mergesort(L.copy())
            end = time.time()
            #Accumulate total time
            mergeSort_times[index] += (end - start)

            #Record times for quicksort
            start = time.time()
            good_sorts.quicksort(L.copy())
            end = time.time()
            #Accumulate total time
            quickSort_times[index] += (end - start)
insertionSort_times = [t / runs for t in insertionSort_times]
mergeSort_times = [t / runs for t in mergeSort_times]
quickSort_times = [t / runs for t in quickSort_times]
```

```python
#Graph of final results displaying list length vs time(s) for each sorting algorithm

plt.plot(length, insertionSort_times, label = "Insertion Sort")
plt.plot(length, mergeSort_times, label ="Merge Sort")
plt.plot(length, quickSort_times, label = "Quick Sort")
plt.xlabel("List Length")
plt.ylabel("Time(s)")
plt.title("List length vs Time(s) for small lists for Insertion, Merge and Quick Sorts")
plt.legend()
plt.show()
```
